**Drop unwanted columns**

```python
#'default' feature(column) does not play important role
df2.groupby(['y','default']).size()
```

```python
#drop 'default'
df2.drop(['default'],axis=1,inplace=True)
```

```python
df2.groupby(['y','pdays']).size()
```

So many negative values

```python
#drop 'pdays'
df2.drop(['pdays'],axis=1,inplace=True)
```

**Outlier remove**

```python
#Outlier in 'age'
df2.groupby('age',sort=True)['age'].count()
#Its distributed normally,outlier can be ignored,values lies in between 18 to 95
```

```python
#Outlier in 'balance'
df2.groupby(['y','balance'],sort=True)['balance'].count()
#when balance goes high ,client shows interest to deposit so outlier can be ignored
```

```python
#Outlier in 'duration'
df2.groupby(['y','duration'],sort=True)['duration'].count()
#when duration goes high ,client shows interest to deposit so outlier can be ignored
```

```python
#Outlier in 'duration'
df2.groupby(['y','campaign'],sort=True)['campaign'].count()
#till 32 campaign clients were interested to deposit,other denied so they can removed as outlier
```

```python
df3=df2[df2['campaign']<33]
```

```python
# pd.reset_option('display.max_rows')
# pd.reset_option('display.max_columns')
df3.groupby(['y','campaign'],sort=True)['campaign'].count()
```

```python
#Outlier in 'duration'
df3.groupby(['y','previous'],sort=True)['previous'].count()
#outlier in previous value jumped from 30 to 55,.. we can remove values above 30 as outlier
```

```python
df4 = df3[df3['previous'] < 31]
```

```python
df4.groupby(['y','previous'],sort=True)['previous'].count()
```

```python
df4.shape
```

```python
df4.sample(6)
```

```python
cat_columns = ['job', 'marital', 'education', 'contact', 'month', 'poutcome']
for col in cat_columns:
    df4_encoded = pd.get_dummies(df4[col], prefix=col, drop_first=True)
    df4 = pd.concat([df4, df4_encoded], axis=1)
    df4.drop(col, axis=1, inplace=True)
```

```python
df4.sample(5)
```

```python
bool_columns=['housing','loan','y']
for col in  bool_columns:
    df4[col+'_new']=df4[col].apply(lambda x : 1 if x == 'yes' else 0)
    df4.drop(col, axis=1, inplace=True)
```

```python
df4.sample(5)
```

#**Balancing Dataset,Split Dataset into Training set and Test set**


```python
x=df4.drop(['y_new'],axis=1)
y=df4['y_new']
```

```python
df4['y_new'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Class Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
from collections import Counter
print("Original Class Distribution:")
print(Counter(y_train))

print("\nResampled Class Distribution:")
print(Counter(y_train_resampled))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_normalized = scaler.fit_transform(x_train_resampled)
x_test_normalized = scaler.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier()
```

```python
parameter={'n_neighbors':[3,5,7,9,11],'weights':['uniform','distance']}
gd=GridSearchCV(clf,parameter,cv=10,scoring='accuracy')
gd.fit(x_train,y_train)
```

```python
model=KNeighborsClassifier(n_neighbors=11,weights='uniform')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```
